# Análise e Agrupamentos das músicas do rock brasileiro

```python
import pickle
import pandas as pd
import numpy as np
import altair as alt
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

import re
from time import time

from sklearn.model_selection import GridSearchCV
```

## Carga dos Dados

```python
letras = pd.read_csv('https://raw.githubusercontent.com/richecr/cdd_dataset/main/letras-ptbr-rock-grande.csv')

letras.sample(10)
```

Contagem de músicas que cada artista possui no dataset utilizado.

Podemos ver que Rita Lee é a artista que possui mais músicas no dataset, um total de 245 músicas.

```python
letras['titulo_limpo'] = letras['SName'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
letras.drop_duplicates(subset ="titulo_limpo", 
                       keep = 'first', inplace = True)

count = letras['Lyric'].str.split().str.len()
letras = letras[count > 50]

print('Letras: {}'.format(letras.shape[0]))
print('\nPor artista: ')
letras['Artist'].value_counts()[0:20]
```

## Pré-processamento dos Dados

Etapas de pré-processamento realizadas:

- Remoção de stop words das letras das músicas.
- Remoção de caracteres especiais e digítos.

```python
stop_words = set(stopwords.words("portuguese"))
stop_words.update(['que', 'até', 'esse', 
                    'essa', 'pro', 'pra',
                    'tudo', 'todo', 'tão', 'bb', 
                    'bbm', 'abm', 'cbm', 
                    'dbm', 'dos', 
                    'ltda', 'editora'])

clean_lyrics = []
for w in range(len(letras.Lyric)):
  lyric = letras['Lyric'].iloc[w]

  lyric  = re.sub("(\\d|\\W)+|\w*\d\w*"," ",lyric )
  lyric = ' '.join(s for s in lyric.split() if (not any(c.isdigit() for c in s)) and len(s) > 2)
  clean_lyrics.append(lyric)

clean_lyrics[0:5]
```

## EDA (Análise Exploratória dos Dados)

Não realizamos nenhuma filtragem, visto que todas as músicas já são do estilo musical que desejamos analisar, rock brasileiro.

Usamos o algoritmo CountVectorizer para transformar a nossa coleção de documentos de texto em uma matriz de contagens de token.

Usamos 1-gram(unigrama) ou 2-gram(bigrama) para geração de nossos tokens.

```python
tf_vectorizer = CountVectorizer(
        min_df = 30,
        max_df = 0.5,
        max_features = 10000,
        stop_words = stop_words, 
        ngram_range = (1,2)
)

vec_text = tf_vectorizer.fit_transform(clean_lyrics)
words = tf_vectorizer.get_feature_names()

print(vec_text.shape)
print(len(words))
```

# Análises

## Encontrando tópicos

O resultado terá 

*   uma matriz que descreve a relação entre palavras e tópicos
*   uma matriz que descreve a relação entre documentos e tópicos

Usamos a implementação do algoritmo LDA usando a biblioteca do python, [Gensim](https://radimrehurek.com/gensim/).

```python
from sklearn.decomposition import LatentDirichletAllocation
```

Funções para uso geral:

print_top_words: Imprimir as n palavras mais frequentes em um determinado tópico.

display_topics: Imprimir as n músicas com maior probabilidade de serem daquele tópico especifico.

```python
def print_top_words(model, feature_names, n_top_words):
  for topic_idx, topic in enumerate(model.components_):
    print("\n--\nTopic #{}: ".format(topic_idx + 1))
    message = ", ".join([feature_names[i]
                          for i in topic.argsort()[:-n_top_words - 1:-1]])
    print(message)
  print()

def display_topics(W, H, feature_names, documents, no_top_words, no_top_documents):
    for topic_idx, topic in enumerate(H):
        print("\n--\nTopic #{}: ".format(topic_idx + 1))
        print(", ".join([feature_names[i]
                for i in topic.argsort()[:-no_top_words - 1:-1]]).upper())
        top_d_idx = np.argsort(W[:,topic_idx])[::-1][0:no_top_documents]
        for d in top_d_idx: 
          doc_data = letras[['Artist', 'SName']].iloc[d]
          print('{} - {} : \t{:.4f}'.format(doc_data[1], doc_data[0], W[d, topic_idx]))
```

## Encontrar o melhor número de tópicos:

Usando o GridSearchCV para rodar com diferentes hiperparâmetros e com isso obtermos os melhores valores para geração do nosso modelo.

Usamos essa célula para carregar o modelo, pois demorava para rodar e achar o melhor número de grupos. Portanto, depois da primeira execução(com resultado satisfatório) nós resolvemos salvar o modelo e ir apenas carregando.

```python
# filename = 'our_model.sav'

# lda = pickle.load(open(filename, 'rb'))
```

```python
search_params = {'n_components': range(3, 7), 'learning_decay': [.5, .7]}

lda = LatentDirichletAllocation()
model = GridSearchCV(lda, param_grid=search_params)
model.fit(vec_text)
```

```python
# Best Model
best_lda_model = model.best_estimator_
print("Best Model's Params: ", model.best_params_)
print("Best Log Likelihood Score: ", model.best_score_)
print("Model Perplexity: ", best_lda_model.perplexity(vec_text))
```

## O Melhor número de grupo:

Pelo algoritmo GridSearchCV o melhor número de grupos é 3, com os hiperparâmetros acima.

```python
lda = LatentDirichletAllocation(batch_size=128, doc_topic_prior=None,
                          evaluate_every=-1, learning_decay=0.5,
                          learning_method='batch', learning_offset=10.0,
                          max_doc_update_iter=100, max_iter=10,
                          mean_change_tol=0.001, n_components=3, n_jobs=None,
                          perp_tol=0.1, random_state=None,
                          topic_word_prior=None, total_samples=1000000.0,
                          verbose=0)

t0 = time()

lda.fit(vec_text)
doc_topic_matrix = lda.transform(vec_text)

print("done in %0.3fs." % (time() - t0))
```

Usamos essa célula para salvar o modelo, pois demorava para rodar e achar o melhor número de grupos. Portanto, depois da primeira execução(com resultado satisfatório) nós resolvemos salvar o modelo e ir apenas carregando.

## Agrupamento usando KMeans

Usando o algoritmo MiniBatchKMeans realizamos um agrupamento dos nosso dados.

```python
from sklearn.cluster import MiniBatchKMeans

kmeans = MiniBatchKMeans(n_clusters=3, init_size=1024, batch_size=2048, random_state=20)
kmeans.fit(doc_topic_matrix)
labels = kmeans.predict(doc_topic_matrix)
```

### Agrupamento de todas as músicas por grupos:

Usando o algoritmo KMeans podemos ver que ainda ficou bastante semelhante a análise feita usando o algoritmo LDA do Gensim.

Também pode ser visto uma boa divisória entre os grupos, mostrando que a clusterização ficou boa. Pode-se argumentar que pudesse dividir em mais grupos, mas como realizamos os testes usando o GridSearchCV e pelo score ele achou o melhor número de grupos sendo 3.

```python
letras2 = letra_topico.assign(grupo = labels, 
                              tsne1 = letras_embedded[:,0], tsne2 = letras_embedded[:,1])

alt.Chart(letras2.sample(500)).mark_circle(
    opacity = .7,
    size = 30
).encode(
    x = 'tsne1',
    y = 'tsne2', 
    color = 'grupo:N',
    tooltip = [str(c) for c in letras2.columns]
).interactive()
```
